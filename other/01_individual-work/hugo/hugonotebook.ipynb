```
import networkx as nx
import matplotlib.pyplot as plt
import csv
```

# Exploring data sets and getting a feeling for what networkx can be used for

```
#Testing networkx and how adding nodes work
G = nx.Graph()
G.add_nodes_from([
    (1, {"color": "blue"}),
    (2, {"color": "yellow"}),
    (3, {"color": "brown"}),
    (4, {"color": "red"}),
    (5, {"color": "green"}),
])

D = nx.Graph()
D.add_nodes_from([
    ('RP'),
    ('RP'),
    ('Sam'),
    ('Watcher')
])
```

```
G.add_edges_from([(1, 2), (1, 3), (4,5)])
D.add_edges_from([('RP', 'Sam')])

```

```
print(G.number_of_nodes())
print(G.number_of_edges())
```

```
subax1 = plt.subplot(121)
nx.draw(D, with_labels=True, font_weight='bold')
#nx.draw_shell(G, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold')
```

```
GoT = nx.Graph()


with open('stormofswords.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print('start')
            line_count += 1
        else:
            GoT.add_node(row[0])
            GoT.add_edge(row[0], row[1])
            #print(row[0], ' ', row[1])
            line_count += 1

        
```

```
subax1 = plt.subplot(121)
nx.draw(GoT, with_labels=True, font_weight='bold')
```
